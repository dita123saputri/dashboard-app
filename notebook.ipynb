### Cleaning Data

```python
# Menghapus jumlah duplicate geolocation_df
geolocation_df.drop_duplicates(inplace=True)
print("Jumlah duplikat geolocation_df = ", geolocation_df.duplicated().sum())
```

```python
# Memperbaiki kesalahan tipe data order_items
order_items_df['shipping_limit_date'] = pd.to_datetime(order_items_df['shipping_limit_date'])
order_items_df.info()
```

```python
# Memperbaiki kesalahan nilai max kolom payment_installments tabel order_items

order_payments_df.loc[order_payments_df['payment_installments'] > 24, 'payment_installments'] = 24
order_payments_df['payment_installments'].max()
```

```python
# Memperbaiki kolom kosong tabel order_reviews

# Lihat nilai paling sering muncul untuk kolom review_comment_title
order_reviews.review_comment_title.value_counts()
# Lihat nilai paling sering muncul untuk kolom review_comment_message
order_reviews.review_comment_message.value_counts()

order_reviews['review_comment_title'].fillna(value="Recomendo", inplace=True)
order_reviews['review_comment_message'].fillna(value="Muito bom", inplace=True)

order_reviews.head()

# Memperbaiki tipe data kolom review_creation_date and review_answer_timestamp

date_time_columns = ['review_creation_date', 'review_answer_timestamp']
for column in date_time_columns:
    order_reviews[column] = pd.to_datetime(order_reviews[column])

order_reviews.info()

order_reviews.head()
```

```python
# Memperbaiki tipe data orders_df
orders_df.head()

date_time_columns = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']
for column in date_time_columns:
    orders_df[column] = pd.to_datetime(orders_df[column])

orders_df.info()
```

```python
# Memperbaiki nilai kosong orders_df

# Interpolarization kolom order_approved_at
orders_df.order_approved_at.interpolate(method='linear', limit_direction='forward', inplace=True)

#Interpolarization kolom order_delivered_carrier_date
orders_df.order_delivered_carrier_date.interpolate(method='linear', limit_direction='forward', inplace=True)

#Interpolarization kolom order_delivered_customer_date
orders_df.order_delivered_customer_date.interpolate(method='linear', limit_direction='forward', inplace=True)

orders_df.isna().sum()
```

```python
# Memperbaiki nilai kosong products_df

#Kolom product_category_name
products_df.product_category_name.value_counts()
products_df.product_category_name.fillna(value="cama_mesa_banho", inplace=True)

#Kolom product_name_lenght
products_df.product_name_lenght.fillna(value=products_df.product_name_lenght.mean(), inplace=True)

#Kolom product_description_lenght
products_df.product_description_lenght.fillna(value=products_df.product_description_lenght.mean(), inplace=True)

#Kolom product_photos_qty
products_df.product_photos_qty.fillna(value=products_df.product_photos_qty.mean(), inplace=True)

# Kolom product_weight_g
products_df = products_df.dropna(subset=['product_weight_g'])

# Kolom product_length_cm
products_df = products_df.dropna(subset=['product_length_cm'])

# Kolom product_height_cm
products_df = products_df.dropna(subset=['product_height_cm'])

# Kolom product_width_cm
products_df = products_df.dropna(subset=['product_width_cm'])

products_df.isna().sum()
```

```python
# Eksplorasi orders_df
orders_df.head()
delivery_time = orders_df['order_delivered_customer_date'] - orders_df['order_delivered_carrier_date']
delivery_time = delivery_time.apply(lambda x:x.total_seconds())

orders_df['delivery_time'] = round(delivery_time/86400)
orders_df.describe()


#Ada outlier, bersihkan outlier

#outliers = [x for x in orders_df['delivery_time'] if x < lower_bound or x > upper_bound]
#print(len(outliers))
#print(outliers)

q25, q75 = np.percentile(orders_df['delivery_time'], [25, 75])
iqr = q75 - q25
lower_bound = q25 - (1.5 * iqr)
upper_bound = q75 + (1.5 * iqr)
orders_df.drop(orders_df[orders_df['delivery_time'] < lower_bound].index, inplace=True)
orders_df.drop(orders_df[orders_df['delivery_time'] > upper_bound].index, inplace=True)

orders_df.describe()

# Ada invalid data, delivered_date minus, data tertukar
#invalid_dates = orders_df[orders_df['order_delivered_customer_date'] < orders_df['order_delivered_carrier_date']]
#print(invalid_dates)

mask = orders_df['order_delivered_customer_date'] < orders_df['order_delivered_carrier_date']

# Tukar nilai pada kolom yang salah
orders_df.loc[mask, ['order_delivered_customer_date', 'order_delivered_carrier_date']] = \
    orders_df.loc[mask, ['order_delivered_carrier_date', 'order_delivered_customer_date']].values

orders_df.describe(include='all')
```

```python
# Ekplorasi product_df, orders_df, dan product_id

orders_orderItems_df = pd.merge (
    left = order_items_df,
    right = orders_df,
    how = "inner",
    left_on = "order_id",
    right_on = "order_id"
)

orders_product_df = pd.merge (
    left = orders_orderItems_df,
    right = product_category_merge,
    how = "inner",
    left_on = "product_id",
    right_on = "product_id"
)

orders_product_df.groupby("product_category_name").size().sort_values(ascending=False)

#orders_product_df.head()

```

```python
# Eksplorasi orders_df dan orders_reviews

orders_reviews_df = pd.merge (
    left = orders_df,
    right = order_reviews,
    how = "inner",
    left_on = "order_id",
    right_on = "order_id"
)

# Ekplorasi customer_df dengan orders_reviews_df

customer_reviews_df = pd.merge (
    left = customers_df,
    right = orders_reviews_df,
    how = "inner",
    left_on = "customer_id",
    right_on = "customer_id"
)

#Eksplorasi order_items_df dan sellers_df

order_sellers_df = pd.merge (
    left = order_items_df,
    right = sellers_df,
    how = "inner",
    left_on = "seller_id",
    right_on = "seller_id"
)

#Ekplorasi product_category_merge dengan order_sellers_df

product_sellers_df = pd.merge (
    left = product_category_merge,
    right = order_sellers_df,
    how = "inner",
    left_on = "product_id",
    right_on = "product_id"
)

#join orders_reviews_df dengan orders_reviews_df

seller_reviews_df = pd.merge (
    left = orders_reviews_df,
    right = order_sellers_df,
    how = "inner",
    left_on = "order_id",
    right_on = "order_id"
)

seller_reviews_summary = seller_reviews_df.groupby(by=["seller_id", "review_score"]).size().reset_index(name="count")
top_rated_sellers=seller_reviews_summary[seller_reviews_summary['review_score'].isin([4, 5])]
top_rated_sellers=top_rated_sellers.groupby(by="seller_id")["count"].sum().reset_index().sort_values(by="count", ascending=False)
```

```python

# Eksplorasi all_df


all_df = pd.merge (
    left = customer_reviews_df,
    right = product_sellers_df,
    how = "inner",
    left_on = "order_id",
    right_on = "order_id"
)

all_df['customer_id'].head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Produk apa yang paling banyak dibeli dan paling sedikit dibeli
topPurchasedCategories = orders_product_df.groupby("product_category_name").size().sort_values(ascending=False).head(5)
leastPurchasedCategories = orders_product_df.groupby("product_category_name").size().sort_values(ascending=True).head(5)

topPurchasedCategories = topPurchasedCategories.reset_index(name='quantity')
leastPurchasedCategories = leastPurchasedCategories.reset_index(name='quantity')

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(x="quantity", y="product_category_name", data=topPurchasedCategories, palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel("Jumlah Produk", fontsize=15)
ax[0].set_title("Produk yang paling banyak dibeli", loc="center", fontsize=15)
ax[0].tick_params(axis='y', labelsize=15)

sns.barplot(x="quantity", y="product_category_name", data=leastPurchasedCategories, palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel("Jumlah Produk", fontsize=15)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Produk yang paling sedikit dibeli", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=15)
plt.show()
```

### Pertanyaan 2:

```python
# Seller mana yang paling banyak mendapat rating produk tinggi
seller_reviews_summary = seller_reviews_df.groupby(by=["seller_id", "review_score"]).size().reset_index(name="count")
top_rated_sellers=seller_reviews_summary[seller_reviews_summary['review_score'].isin([4, 5])]
top_rated_sellers=top_rated_sellers.groupby(by="seller_id")["count"].sum().reset_index().sort_values(by="count", ascending=False).head(10)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 10))

colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
sns.barplot(x="seller_id", y="count", data=top_rated_sellers, palette=colors, ax=ax)
ax.set_ylabel(None)
ax.set_xlabel("Seller ID", fontsize=15)
ax.set_title("Seller yang paling banyak mendapat rating produk tinggi", loc="center", fontsize=15)
ax.tick_params(axis='x', labelsize=15)

plt.xticks(rotation=45, ha="right", fontsize=10)
plt.tight_layout()
plt.show()
```

```python
#RFM Analysis
rfm_df = all_df.groupby(by="customer_id", as_index=False).agg({
    "order_id": "nunique",
    "order_purchase_timestamp": "max",
    "price": "sum"
})

rfm_df.columns =["customer_id", "frequency", "max_order_date", "monetary"]

#Kapan pelanggan melakukan transaksi
rfm_df["max_order_date"] = pd.to_datetime(rfm_df["max_order_date"])
rfm_df["max_order_date"] = rfm_df["max_order_date"].dt.date
recent_date = orders_df["order_purchase_timestamp"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_date"].apply(lambda x: (recent_date - x).days)

rfm_df.drop("max_order_date", axis=1, inplace=True)
rfm_df.head()

# Visualisasikan rfm
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(24, 8))

colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4"]

sns.barplot(y="recency", x="customer_id", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency (Days)", loc="center", fontsize=18)
ax[0].tick_params(axis='x', labelsize=15)
ax[0].tick_params(axis='x', rotation=45)

sns.barplot(y="frequency", x="customer_id", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)
ax[1].tick_params(axis='x', rotation=45)

sns.barplot(y="monetary", x="customer_id", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)
ax[2].tick_params(axis='x', rotation=45)

plt.suptitle("Best Customers Based on RFM Parameters (customer_id)", fontsize=20)
plt.tight_layout()
plt.show()
rfm_df.head()
```
